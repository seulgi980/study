### TreebankWordTokenizer
- 표준토크나이저

```python
from nltk.tokenize import TreebankWordTokenizer

# 함수가아니고 class이기때문에 객체를 만들고 시작
tokenizer = TreebankWordTokenizer()
tokenized_word = tokenizer.tokenize(text)
tokenized_word
```

```python
# 문장 정제시 각 문장의 단어의 원형(lemma)을 구해서 다시 문장으로 재구성
# 1. 각 단어의 lemma구하기
# 2. lemma를 띄어쓰기 기준으로 다시 합쳐주기

import nltk
nltk.download("punkt")
nltk.download("wordnet")
nltk.download('omw-1.4')

from nltk.stem.wordnet import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# 컴프리헨션쓰는게 좋음
word_lemmas = []
for word in tokenized_word:
  word_lemma = lemmatizer.lemmatize(word.lower(), "v") # lower 소문자로바꾸기
  word_lemmas.append(word_lemma)

print(word_lemmas)
```

```python
# 다시합치기
print(" ".join(word_lemmas))
```

```python
# spacy써주면 완전편함
import spacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
word_lemmas = [ word.lemma_ for word in doc ]
word_lemmas
```

```python
' '.join(word_lemmas)
```

### Okt, 꼬꼬마(Kkma), 코모란(Komoran), 한나눔(Hannanum), Mecab
* 파이썬의 `Duck Typing` 형식으로 사용하는 메소드가 거의 통일되어 있다.
  * `nouns` : 명사만 추출할 때 사용
  * `morphs` : 각 형태소 별 토큰화(명사, 동사, 형용사 등등..)
  * `pos` : 각 형태소 토큰 및 형태소 종류를 튜플로 표현

```python
# Duck Typing 문법 예시
# Duck Typing : 메소드의 이름이 클래스의 정체성을 결정해주는 문법

class A:
  def foo(self):
    print("here is A")

class B:
  def foo(self):
    print("here is B")

class C:
  def goo(self):
    print("here is C")

# A랑 B가 같은 foo메소드를 가지고있으니까 같은 클래스로 취급하는거.
# C는 다른클래스
# 내새끼가 아니면 안따라감. 오리

# ABC = 새끼오리
# AB는 같은 어미에서 낳은새끼
# C는 다른새끼

# 어미오리의 역할을 하는 메소드(함수, 클래스)
def print_foo(obj):
  obj.foo()

print_foo(A())
print_foo(B())
# C만 못따라감.
```

```python
from konlpy.tag import Hannanum, Kkma, Komoran, Okt

hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
okt = Okt()

sentence = "좋으니 그 사람 솔직히 견디기 버거워"
```

```python
def print_tokens(tokenizer, s): # s는 문장 sentence
  print(tokenizer.nouns(s)) # nouns : 명사만 추출
  print(tokenizer.morphs(s)) # 각 형태소 별로 토큰화
  print(tokenizer.pos(s)) # part of speech 각 형태소 토큰 및 종류를 튜플로 표현
```

#### 한나눔(Hannanum)

```python
print_tokens(hannanum, sentence)
```

# Keras의 Tokenizer를 이용해서 정수 인코딩 하기
- 정수인코딩은 단어에 대해 수행

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
```

```python
# 문장들이 모여있기 때문에 문서라고함.

text = """Isn't she lovely.
Isn't she wonderful.
Isn't she precious.
Less than one minute old.
I never thought through love we'd be.
Making one as lovely as she.
But isn't she lovely made from love.
Isn't she pretty.
Truly the angel's best.
Boy, I'm so happy.
We have been heaven blessed.
I can't believe what God has done.
Through us he's given life to one.
But isn't she lovely made from love.
Isn't she lovely.
Life and love are the same.
Life is Aisha.
The meaning of her name.
Londie, it could have not been done.
Without you who conceived the one.
That's so very lovely made from love."""
```

```python
# 문서를 문장으로 만들어주기
sent_tokens = sent_tokenize(text)
print(sent_tokens)
```

```python
# fit_on_texts() 메소드에 문장 코퍼스(말뭉치)를 넣어주면 단어 빈도수를 기준으로 단어 vocalurary(집합)를 만들어줌!
# 빈도수가 많을수록 앞의번호
tokenizer.fit_on_texts(sent_tokens)
```

```python
# 단어 집합 확인
tokenizer.word_index
```

```python
# 단어의 빈도수 확인
tokenizer.word_counts
```

```python
# 인코딩 : 텍스트를 > 숫자화
# texts_to_sequences 메소드 사용

# 단어 코퍼스의 형태로 들어가야 하기 때문에 2차원 배열 형태로 넣어줘야
tokenizer.texts_to_sequences([word_tokenize('she is lovely')])
```

```python
tokenizer.texts_to_sequences(['she is lovely', 'life is lovely']) # 그냥 문장으로 넣어줘도 되나봄
```

```python
# 디코딩( 숫자 > 텍스트화 )
tokenizer.sequences_to_texts([[1, 51, 3], [9, 51, 3]])
```

```python
tokenizer.sequences_to_texts([[1,5,6,31,50,56]])
```

# OOV(out of vocabulary)설정

```python
# 빈도수 상위 5개의 토큰만 사용하도록 할 수도 있다.
vocab_size = 5

# num_words사용
tokenizer = Tokenizer(num_words= vocab_size + 2, oov_token='<oov>') # +2한 이유. pad와 oov까지 포함.
tokenizer.fit_on_texts(sent_tokens)
```

```python
tokenizer.texts_to_sequences(['she is bowwow'])
```

```python
tokenizer.word_index
# 나머지도 만들어는지는데 상위 5개만쓰임
```

# Padding

```python
integer_tokens = tokenizer.texts_to_sequences(sent_tokens)
integer_tokens
# 문장 다 정수로 바꾼거
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 기본은 prepadding(앞쪽에 채워주는거, 앞쪽짤림)
padded_tokens = pad_sequences(integer_tokens, maxlen=5) # 모든문장의 길이가 5가됨 ex)7짜리는 짤림
padded_tokens
```

```python
padded_tokens = pad_sequences(integer_tokens, padding='post', maxlen=5)
padded_tokens
```

# 어느부분을 자르고 어느부분을 채울지 옵션으로 선택가능
- padding = 'post or pre' : 앞쪽에 패딩채울건지, 뒷쪽에 채울건지
- truncating = 'post or pre' : 앞쪽을 자를건지 뒤쪽을 자를건지
