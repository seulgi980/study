## TF-IDF (Term Frequency - Inverse Document Frequency)
DTM을 기반으로 중요한 단어에 가중치는 주는 방식이에요! 결과적으로만 말하자면 TF-IDF의 값을 기반으로 중요한 단어는 값이 올라가고, TF-IDF 기준으로 중요하지 않은 단어는 값이 DOWN이 됩니다!

tf-idf의 정의에 대해 이야기 해보겠습니다.

* $tf(d, t)$ : 특정 문서 d에서의 특정 단어 $t$의 등장 횟수. 즉 DTM 상에서의 단어들의 값
* $df(t)$ : 단어 $t$가 등장한 문서의 수
* $idf(t)$ : $df(t)$에 반비례 하는 수.

참고로 idf는 다음과 같아요!

$$
idf(t) = log(\frac{n}{1+df(t)})
$$

위 식에서 $n$은 문서의 개수입니다!

```python
# 원래는 전처리로 소문자화도 해줘야함
test = []
for c in range(len(corpus)):
  r = corpus[c].lower()
  test.append(r)

test
# corpus 소문자로바꾼거
```

```python
# n번 문서(document)에 단어(term)가 등장한 횟수

def term_frequency(term, document):
  return document[0].count(term) # [0]번 넣어줘야 나오는듯

# def term_frequency(term, document):
#   cnt_vector = CountVectorizer()
#   cnt_vector.fit(document)
#   tf = cnt_vector.transform(document).toarray()
#   idx = cnt_vector.vocabulary_[term]
#   result = []
#   for i in range(len(document)):
#     result.append(tf[i][idx])
#   return result



# 단어(term)가 몇개의 문서'들'(documents)에서 등장했는지 횟수
def document_frequency(term, documents):
  count = 0
  for doc in documents:
    if term in doc:
      count += 1

  return count


def inverse_document_frequency(term, documents):
  from math import log
  df = document_frequency(term, documents)

  return log(len(documents) / (1 + df))




def tf_idf(term, documents, n): # 몇 번째 문서

  tf =  term_frequency(term, documents[n-1])
  idf = inverse_document_frequency(term, documents)

  return tf * idf
```

# 20 뉴스그룹 분류

```python
from sklearn.datasets import fetch_20newsgroups

news_data = fetch_20newsgroups(subset='all', random_state=42)
```

```python
# target 이름 확인
news_data.target_names

# 뉴스 카테고리들
```

```python
# 데이터 확인
print(news_data['data'][0])
```

```python
# 훈련 데이터 가져오기. header, footers, 특수문자를 제거하고 불러오기
train_news = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'), random_state=42)
X_train = train_news['data']
y_train = train_news['target']

test_news = fetch_20newsgroups(subset='test', remove=('headers', 'footers', 'quotes'), random_state=42)
X_test = test_news['data']
y_test = test_news['target']
```

```python
X_train[:3]
# 뉴스기사 3개만봐보기 뉴스기사 corpus가 됨

# preprocessing필수!
```

```python
# Text Feature를 벡터화 시켜보기
# CountVectorizer나 TfidfVectorizer 사용

cnt_vector = CountVectorizer()
X_train_cnt_vect = cnt_vector.fit_transform(X_train)
X_test_cnt_vect = cnt_vector.transform(X_test) # 검증세트는 fit안함!
```

```python
X_train_cnt_vect.shape
# 대략 10만개의 단어들을 사용하고있군
# 뉴스의양은 11314개
# DTM 가로세로 잘알기!
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_clf = LogisticRegression(solver='liblinear')

lr_clf.fit(X_train_cnt_vect, y_train)
pred = lr_clf.predict(X_test_cnt_vect)

accuracy_score(y_test, pred)
```

```python
# 예측할때는 문장, 문서를 ⭐벡터화⭐해서 넣어야함.


predict_news_vect = cnt_vector.transform(['아무기사~~~']) # 리스트 안쪽에 넣기
lr_clf.predict(predict_news_vect)
```
