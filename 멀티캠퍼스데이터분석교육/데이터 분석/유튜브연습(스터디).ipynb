```python
import pandas as pd

KRVideos = pd.read_csv('/content/drive/MyDrive/멀티캠퍼스 9회차 I반/05데이터 분석/KR_youtube_trending_data.csv')

KRVideos.head()
```

```python
cols = ['title','channelTitle', 'view_count','likes']
df = KRVideos[cols]
df.info()
```

```python
df_drop = df.drop_duplicates(['title','channelTitle'], keep='first')
df_100 = df_drop[ :100]
df_100.sort_values(by='view_count', ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.scatter(df_100['view_count'],df_100['likes'], s=500)  # s는 점크기
plt.xlabel('조회수')
plt.ylabel('좋아요수')

plt.grid()
plt.show()
```

```python
import numpy as np
fp1 = np.polyfit(df_100['view_count'],df_100['likes'],1)
fp1
```

```python
f1 = np.poly1d(fp1)

fx = np.linspace(0, 80000000, 10000)

# 그래프 사이즈 조절
plt.figure(figsize=(10, 10))

plt.scatter(df_100['view_count'],df_100['likes'], s=50)

line = f1(fx)
plt.plot(fx, line, color='g', linestyle='--')

plt.xlabel("조회수")
plt.ylabel("좋아요수")
plt.grid()
plt.show()
```
