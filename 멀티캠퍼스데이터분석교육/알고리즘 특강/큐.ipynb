```python
def isQueueFull():
  global SIZE, queue, front, rear
  if (rear >= SIZE -1):
    return True

  else:
    return False

```

```python
def enQueue(data):
  global SIZE, queue, front, rear
  if isQueueFull() == False:
    rear += 1
    queue[rear] = data

    print('<<---출구', queue, '<<---입구')

  else:
    print("꽉차서못너")
    return None
```

# 미리 맨앞까지땡겨놓지말고. 1명들어오면 1칸씩 땡기는 방식으로

image.png

```python
def isQueueFull():
  global SIZE, queue, front, rear
  if (rear != SIZE -1): # case1 뒤에 여유자리 남은경우
    return False


  elif (rear == SIZE -1 and front == -1):  # case2 빈틈없이 꽉찬경우
    return True



  else: # case3 rear는 꽉차고 앞에가 비어있는경우. 1칸씩만 땡깁시다⭐
    for i in range(front+1,SIZE):
      queue[i-1] = queue[i] # 1칸앞으로옮기고
      queue[i] = None # 원래데이터 지워서 문별 땡겨!
    front -= 1
    rear -= 1
    return False

```

# 원형큐는 1칸 못씀 왜?
# front == rear면 큐가 빈거⭐⭐⭐⭐⭐⭐

그러므로 다찬건지 빈건지 구분을못함

```python
# 원형큐 초기화
SIZE = 5
queue = [None for _ in range(SIZE)]
front, rear = 0, 0
```

```python
## 함수
def isQueueFull() :
    global SIZE, queue, front, rear
    if ( (rear+1)  % SIZE == front) :
        return True
    else :
        return False

def enQueue(data) :
    global SIZE, queue, front, rear
    if(isQueueFull()) :
        print('큐 꽉!')
        return
    rear = (rear + 1) % SIZE
    queue[rear] = data

def isQueueEmpty() :
    global SIZE, queue, front, rear
    if (front == rear) :
        return True
    else :
        return False

def deQueue() :
    global SIZE, queue, front, rear
    if (isQueueEmpty()) :
        print('큐 텅~')
        return None
    front = (front + 1) % SIZE
    data = queue[front]
    queue[front] = None
    return data

def peek() :
    global SIZE, queue, front, rear
    if (isQueueEmpty()) :
        print('큐 텅~')
        return None
    return queue[(front+1) % SIZE]

```

```python
enQueue('화사')
enQueue('솔라')
enQueue('문별')
enQueue('휘인')

print('출구<---', queue, '<---입구')
```

```python
retData = deQueue()
print('*식사손님:', retData)
print('**다음 준비 손님:', peek())
retData = deQueue()
print('*식사손님:', retData)

print('출구<---', queue, '<---입구')
```
