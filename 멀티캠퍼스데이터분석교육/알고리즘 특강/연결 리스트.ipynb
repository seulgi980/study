```python
def PrintNodes(start):
  current = start
  print(current.data, end=' ')

  while (current.link != None) :
    current = current.link
    print(current.data, end=' ')

  print()



memory = []
head, current, pre = None, None, None
dataArray = ['다현', '정연','쯔위', '사나', '지효']

# 연결리스트 생성⭐

node = Node()
node.data = dataArray[0]
head = node # 첫번째노드
memory.append(node)


for data in dataArray[1:]:
  pre = node  # node를 재사용할거라서 pre에 잡고있게함
  node = Node() # node재활용 새로운node만듦
  node.data = data
  memory.append(node)
  pre.link = node
```

```python
PrintNodes(head)
```

```python
# ⭐⭐⭐⭐⭐⭐⭐
def insertNode(findData, insertData) :
    global memory, head, current, pre
    ## Case 1 : 헤드 앞에 삽입할 경우 (다현, 화사)
    if (findData == head.data) :
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        memory.append(node)
        return
    ## Case 2 : 중간 노드 앞에 삽입할 경우 (사나, 솔라)
    current = head
    while (current.link != None) :
        pre = current
        current = current.link
        if (current.data == findData) :
            node = Node()
            node.data = insertData
            node.link = current
            pre.link = node
            memory.append(node)
            return
            
    ## Case 3 : 없는 노드 앞에 삽입할 경우(== 제일 뒤에 추가) (재남, 문별)
    node = Node()
    node.data = insertData
    current.link = node
    memory.append(node)
    return

## 전역 변수
memory=[]
head, current, pre = None, None, None
dataArray = ['다현', '정연', '쯔위', '사나', '지효'] # 나만의 데이터

## 메인

## 연결리스트 생성
node = Node()
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:] : # ['정연', '쯔위', ....
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)

printNodes(head)

## 연결 리스트 삽입(*)
# insertNode('다현', '화사') # 찾을데이터, 삽입할데이터
# insertNode('사나', '솔라')
insertNode('재남', '문별')
printNodes(head)
```

```python
# 연결리스트 삭제⭐
def deleteNode(delData) :
    global memory, head, current, pre
    ## Case 1 : Head를 삭제할 때. (다현)
    if (delData == head.data) :
        current = head
        head = head.link
        del(current)
        return

    ## Case 2 : 중간노드를 삭제할 때. (쯔위)
    current = head
    while (current.link != None) :
        pre = current
        current = current.link
        if (current.data == delData) :
            pre.link = current.link
            del(current)
            return
            
    ## Case 3 : 없는 노드를 삭제할 때. (재남)
    return
```
