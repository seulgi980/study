```python
stack = [None, None, None, None, None]
top = -1

print("바닥 : ", stack)
```

```python
# Push 데이터 삽입
def Push(data):
  global top, stack, SIZE
  if IsStackFull() == True:
    print("꽉차서 못넣엉")
    return

  else:
    top += 1
    stack[top] = data

  print("바닥 : ", stack)

```

```python
# Pop 데이터 빼기
def Pop():
  global top, stack, SIZE

  if (IsStackEmpty()) :
    print('스택 텅~')
    return None

  else:
    data = stack[top]
    stack[top] = None
    top -= 1

    print("pop한 데이터 : {}".format(data))
    print("바닥 : {}".format(stack))
```

```python
# stack n개짜리만들기
SIZE = 5
stack = [None for _ in range(SIZE)]

stack
```

```python
# 스택이 꽉찼을수도있으니 확인하는 함수
def IsStackFull():
  global top, stack, SIZE
  if (top >= SIZE - 1):
    return True


  else:
    return False
  
```

```python
def IsStackEmpty() :
  global SIZE, stack, top
  if (top <= -1) :
    return True
  else :
    return False
```

```python
# top위치의 데이터를 확인하는 함수
def peek() :
    global SIZE, stack, top
    if (IsStackEmpty()) :
        print('스택 텅~')
        return None
    return stack[top]
```
