```python
numData = [[ 5, 7, -5, 100, 73 ],
          [ 35, 23, 4, 190, 33 ],
          [ 49, 85, 662, 39, 81 ],
          [ 124, -59, 86, 46, 52 ],
          [ 27, 7, 8, 33, -56 ]]
```

```python
def replaceData(numData): 
  retData = [] # 조건에 따라서 전처리된 2차원 배열
  
  for i in range(0,5):
    a = []
    for j in range(0,5):
      if numData[i][j] < 0:
        a.append(0)

      elif numData[i][j] > 100:
        a.append(numData[i][j] % 100)

      else:
        a.append(numData[i][j])
      
    retData.append(a)


  return retData
```

```python
retData = replaceData(numData)
retData
```

```python
def getMaxSum(retData):
  maxSum = 0
  for j in range(0,4):
    for i in range(0,4):
      sum = (retData[i][j] + retData[i+1][j] + retData[i][j+1] + retData[i+1][j+1])
      if sum >= maxSum:
        maxSum = sum
      else:
        maxSum = maxSum


  return maxSum
  

```

```python
getMaxSum(retData)
```
