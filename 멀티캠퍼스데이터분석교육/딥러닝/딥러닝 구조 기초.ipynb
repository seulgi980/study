# 활성화 함수

$$
\sigma(x) = \frac{1}{1+exp(-x)}
$$

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))


def identity_function(x):
  return x
```

image.png

```python
# 문제!! 1층의 활성화함수는 sigmoid
# 2층=출력층. 활성화함수는 항등함수로 저모양 신경망짜기

X= np.array([0.7, 1.8])

W= np.random.rand(2,3)
B= np.random.rand(3)

print(X)
print(W) # 1층의 가중치
print(B) # 1층의 편향
```

```python
Z = X.T@W + B
A = sigmoid(Z) # 1층의 활성화함수적용
A
```

```python
W2 = np.random.rand(3,2) # 2층의 가중치
B2 = np.random.rand(2) # 2층의 편향

Z2 = A@W2 + B2
Y = identity_function(Z2) # 2층의 활성화함수 적용
Y
```
