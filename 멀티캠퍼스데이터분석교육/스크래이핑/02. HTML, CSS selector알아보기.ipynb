# HTML(html 은 웹 브라우저에서 전체적인틀, 뷰를 담당하는 마크업 언어입니다)
 - 골격을 만듦
 - HyperText Markup Language

  HyperText : 페이지의 이동이 가능한 (원하는 부분을 이곳저곳 열어갈 수 있다. )

  Markup : 구간이있는(표식⭐⭐)

  Language : 언어

- Markup= tag등을 활용해 문서의 골격,구조를 정의⭐

  나는 어제 사과를 먹었다 컴퓨터를 못알아들음
    -  ```html
        <누가>나</누가>
        <언제>어제</언제>
        <무엇을>사과</무엇을>
        <어떻게>먹었다</어떻게>
    ```

- HTML은 웹에 보이는 화면인 웹 페이지에 필요한 구성요소를 구성하는 Markup 언어
- 웹 페이지를 구성하는 언어
  - HTML : 마크업을 이용해 화면의 레이아웃 구조와 텍스트, 엘리먼트를 표현한다.
  - CSS(Cascade Style Sheet) : 엘리먼트를 꾸며주는 언어
    - **html 엘리먼트를 선택**을 한 다음 꾸며줘야 한다.
      - `selector`를 이용한 엘리먼트 선택
  - JavaScript : 사용자와 웹 페이지의 인터렉션을 담당

## HTML의 구성 요소
- Document
  - 눈에 보이는 페이지 전체
- Element
  - Document를 구성하는 **요소**
  - 레이아웃(칸나누기 방의 벽) 일반적으로 눈에 보이진 않음 (네이버예시)
    - 영역을 나눠주는 역할. 대표적으로 `<div>, <p>, <span> 등등`
  - 엘리먼트
    - 특정한 기능이나 표현이 있는 구성요소(네이버 검색창: 검색기능, 아이콘: 누르면 어떤 기능, 로고 : 표시표현)
    - 눈으로 볼 수 있는 요소
    - 레이아웃이 아니면 모두 엘리먼트
  - 엘리먼트는 **계층적 구조**로 이루어져 있다.⭐⭐⭐
    - 엘리먼트 내에 다른 엘리먼트가 배치될 수 있다.
    - **DOM(Document Object Model)** ★★★★★
- Tag(mark up에서)
  - 목적이 다른 태그들이 각각 존재한다.
  - 태그의 이름에 따라서 태그의 모양과 기능(목적)이 달라진다.

  - <안녕> text </안녕>  안녕기능 태그
  - 태그들을 이용해 엘리먼트 구조를 만들어 낸다.
- Attribute (클래스의 멤버변수와비슷)
  - 엘리먼트의 속성을 표현 `attribute="attribute value"`
  - 시작 태그 안에 포함되는 속성값

## 속성값 종류
- `id, class` 속성⭐⭐⭐
  - **엘리먼트를 선택**하기 위한 속성
  - `id`속성 : 전체 웹 페이지에서 유일한 값(기본키느낌) 결과물이 element로나옴
  - `class`속성
    - 웹 페이지에서 여러 가지 값을 사용할 수 있다. **중복 가능** 결과물이 list로 나온다.
    - 하나의 엘리먼트 내에 여러 개의 `class`를 설정 가능
- `기타 등등`
  - 엘리먼트의 기능 등을 담당
  - 엘리먼트의 크기, 배경 화면 등등 스타일 적인 요소를 담당
  - `a` 태그 같은 경우는 이동할 페이지를 입력


# `table` 태그
- 표를 그리기 위한 태그 (데이터 프레임)
    * `table` : 전체 테이블을 만들기 위한 태그
    * `caption` : 표의 제목을 만들기위한 태그(생략 가능)
    * `thead` : 테이블을 항목(컬럼)을 표시하기위한 태그 (생략 가능)
    * `tbody` : 테이블에 표시할 실제 데이터가 들어가는 태그 (생략 가능)
    * `th` : 컬럼명을 표기하기 위한 태그
    * `td` : 데이터를 표기하기 위한 태그
    * `tr` : 한 행을 만들기 위한 태그

# CSS Selector
- CSS : html 엘리먼트를 꾸미기 위한 시트
  - 엘리먼트를 선택자(selector)를 이용해 선택을 하고 꾸미게 됩니다.
- CSS 생성 방법
  - HTML 내에 `<style></style>`
  - 엘리먼트의 style attribute를 활용
    - `style="color:red; background=blue;"`
    - 선택자를 활용하지 않는 방법.

## Selector의 종류
- 태그의 이름을 이용해서 선택
  - `div, span, input, p, a, ul, li 등등..` 태그명을 이용해서 선택
  - 한 웹페이지에 여러 개의 태그가 있을 수 있기 때문에 **여러 개**의 엘리먼트가 선택
- `id` 속성을 이용해서 선택
  - 웹 페이지에 존재하는 유일한 속성을 이용해 엘리먼트를 선택
  - 1개의 엘리먼트만 선택
- `class` 속성을 이용해서 선택
  - 웹 페이지에 여러 개 존재하는 값을 이용해서 선택
  - **여러 개**의 엘리먼트를 선택
- `attribute` 속성을 이용해서 선택
  - 엘리먼트에 특정 효과를 부여하는 `attribute`를 이용해서 여러 개의 엘리먼트를 선택

## CSS 문법
```html
<style>
  selector {
    꾸미기 시트
  }
</style>
```

# 계층 구조 엘리먼트 선택⭐⭐⭐⭐
- DOM(document object model)을 활용해서 선택

자식은 부모 > 자식 (꺾쇠 괄호)

자손은 부모 자식  (띄어쓰기)


````python
%%html

<div class="container">
  <h3 class="title">inner 1</h3>
  <div class="wrapper">
    <h4 class="title">inner 2</h4>
  </div>
</div>
````

image.png

container의 후손은 3개!

````python
%%html

<style>
  .container > .title {
    color: red;
  }

  .container .title{
    background-color: yellow;
  }

</style>


<div class="container">
  <h3 class="title">inner 1</h3>
  <div class="wrapper">
    <h4 class="title">inner 2</h4>
  </div>
</div>
````
