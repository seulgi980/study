```python
# 데이터프레임만들기
import pandas as pd

fin_list = soup.select("#searchList > li")



brand_list = []
title_list = []
goods_link_list = []
price_list = []
bn_list = []
season_list = []
female_list = []

result_img_list = []


for fin in fin_list:
  brand = fin.select_one(".item_title > a").text
  title = fin.select_one(".list_info").text.strip()
  goods_link = fin.select_one(".img-block").get("href").split("//")[-1]
  price = fin.select_one(".price").text.split()[-1]


  brand_list.append(brand)
  title_list.append(title)
  goods_link_list.append(goods_link)
  price_list.append(price)


  product_code = fin.select_one(".img-block").get("href").split("/")[-1]
  URL2 = f"https://www.musinsa.com/app/goods/{product_code}"
  
  
  headers = {
    "user-agent": UserAgent().chrome
  
  }
  response = requests.get(URL2, headers=headers)

  page2 = response.content
  soup2 = BeautifulSoup(page2, 'html.parser')

  try:
    bn = soup2.select_one("p.product_article_contents > strong").text
  except:
    bn = "✔✔✔"

  try:
    season = ",".join(soup2.select("p.product_article_contents > strong")[1].text.split())
  except:
    season = "✔✔✔"

  try:
    female = "😁".join(soup2.find("span", {"class":"txt_gender"}).text.split())
  except:
    female= "✔✔✔"



  brand_list.append(bn)
  season_list.append(season)
  female_list.append(female)



  result_img = []

  for i in soup2.select(".product_thumb > li > img"):
    b= i["src"]
    result_img.append(b)

  result_img_list.append(result_img)

brand_list, title_list, goods_link_list, price_list, bn_list, season_list,female_list, result_img_list
```

```python
result = pd.DataFrame({
    "브랜드": brand_list,
    "상품명" : title_list,
    "상품상세페이지주소": goods_link_list,
    "가격": price_list,
    "품번" : bn_list,
    "시즌" : season_list,
    "성별" : female_list,
    "이미지주소들" : result_img_list,
    
    })

result.loc[20]
```
