```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://www.musinsa.com/categories/item/001005"
```

```python
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
products = soup.select("#searchList > li")
len(products)
```

```python
# 첫번째 상품 가져오기
sample = products[0]
sample
```

```python
sample_image = sample.select_one("img.lazyload")
sample_image
```

```python
img_url = sample_image['data-original']
img_url
```

```python
img_url = "https:" + img_url
img_url
```

```python
# 이미지 주소로 들어가서 다운받아야
from urllib.request import urlretrieve

image_name = img_url.split("/")[-1]

urlretrieve(img_url, image_name) # urlretrieve(이미지주소, 저장할경로)
```

```python
# 구글 드라이브에 다운받아보기
# 위에껀 그냥 왼쪽 파일에 바로들어감
urlretrieve(img_url, '/content/drive/MyDrive/멀티캠퍼스 9회차 I반/스크래이핑/' + '감자' )
```
