## 다중 그래프
- 1개의 캔버스 안에 여러개의 그래프 그리기

```python
data1 = np.arange(1,51)
data2 = np.arange(51,101)

plt.plot(data1)
plt.plot(data2)

plt.show()


# 여러개 그래프 동시에 올리기도 가능
```

```python
data1 = np.arange(1,51)
data2 = np.arange(51,101)

# plt.figure() 하나는 자동으로 만들어져서 굳이 안써도됨
plt.plot(data1)

plt.figure()
plt.plot(data2)

plt.show()

# 캔버스가 2개인형태
```

# 하나의 캔버스에 여러개의 plot을 그리기 subplot
  - plt.subplot(row,column, index)

```python
plt.subplot(2,1,   1) # 하나의 캠퍼스를 2X1로 쪼개!
plt.plot(data1)


plt.subplot(2,1,   2)  # 인덱스 1부터시작
plt.plot(data2)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(data1)

plt.subplot(3,1,2)  # 숫자달라지면 위에꺼무시됨
plt.plot(data2)

plt.show()

# 그래서 행열은 한번정하면 쭉가야!!
```

```python
plt.subplot(1,2,1)
plt.plot(data1)

plt.subplot(1,2,2)
plt.plot(data2)

plt.show()
```

```python
fig, axes = plt.subplots(2, 3)
axes_ravel = axes.ravel()

# for문써도됨
for ax in axes_ravel:
  ax.plot(data1)

plt.show()
```

```python
# 응용
fig, axes = plt.subplots(2, 3)
axes_ravel = axes.ravel()
datas = [data1, data1 * data1, data1 ** 3, data % 10, data2 // 10, data2 ** 2]

for ax, data in zip(axes_ravel, datas):
  ax.plot(data)

plt.show()
```

# 스타일 세부설정

## marker

**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker

```python
plt.plot(np.arange(10), np.arange(10) * 2, marker='o', markersize=5)
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', markersize=30)
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', markersize=15)
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', markersize=20)

# 타이틀 & font 설정
plt.title("마커 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

## linestyle

**line의 종류**
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', linestyle='')   # 선이 안보임
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', linestyle='-')  # 실선
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', linestyle='--') # 짧은 실선
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', linestyle='-.') # 선과 점을 엮어서
plt.plot(np.arange(10), np.arange(10) * 2 - 40, marker='v', linestyle=':')  # 점섬

# 타이틀 & font 설정
plt.title("라인 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.show()
```

# color

**color의 종류**
* 'b'	blue
* 'g'	green
* 'r'	red
* 'c'	cyan
* 'm'	magenta
* 'y'	yellow
* 'k'	black
* 'w'	white

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', linestyle='', color='b')
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', linestyle='-', color='r')
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', linestyle='--', color='c')
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', linestyle='-.', color='y')

# 타이틀 & font 설정
plt.title("색상 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

# alpha

투명도 설정 0~1사이즈로 조절. 값이 작을수록 투명해진다.

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', color='b', alpha=0.1)
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', color='r', alpha=0.3)
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', color='c', alpha=0.6)
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', color='y', alpha=1.0)

# 타이틀 & font 설정
plt.title("투명도( alpha ) 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```
