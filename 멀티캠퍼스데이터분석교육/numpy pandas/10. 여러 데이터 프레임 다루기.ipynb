# DataFrame 합치기! concat⭐

## row 방향 합치기
- axis = 0
- 순서 뒤죽박죽이어도 컬럼명이 같은 것들끼리 합쳐짐!!⭐⭐
- ㅁ
- ㅁ ↓
- ㅁ 아래쪽으로 붙이기

```python
df1_copy = df1.copy()
df1_concat = pd.concat([df1,df1_copy], axis = 0)  # 축 0 번방향으로 붙이기!!
df1_concat
```

```python
# 행인덱스가 0~14 0~14인게 조금 마음에 안드네
# 기존에 있던 인덱스를 사용하지 않고 재구성

df1_concat.reset_index()

# 기존 index도 남음!
```

```python
df1_concat.reset_index(drop=True)
# 기존 인덱스도 삭제해서 재구성!
```

```python
pd.concat([df1,df2],axis=0)
```

# column 방향 합치기
- axis =1
- 순서 섞였어도 행 인덱스가 일치하는 것 끼리 합쳐짐!⭐⭐

```python
pd.concat([df1,df2], axis =1)
```

```python
df2_sorted = df2.sort_values("연봉")
# 정렬해서 인덱스 뒤죽박죽일 때 합치면?

pd.concat([df1,df2_sorted], axis=1)

# index기준으로 합치기때문에⭐⭐⭐ 정상적으로 됨!!
```

```python
df2_sorted = df2.sort_values("연봉")
df2_reset = df2_sorted.reset_index(drop=True)
df2_reset

# 그럼 인덱스자체를 바꿔버리면?

```

```python
pd.concat([df1,df2_reset], axis=1)

# 인덱스 기준으로 합치기때문에 데이터가 안맞다!!⭐⭐⭐
```

# DataFrame 병합
- concat : 단순하게 row나 column방향으로 이어 붙이기.
   인덱스 기준⭐⭐⭐으로 이어붙여진다!!

- merge : 특정한 고유키(unique id)값을 기준으로 병합!
  이름같은 고유키로!!

```python
df2_right = df2.drop([1,3,5,7,9])
df2_right = df2_right.reset_index(drop=True)
df2_right

# 5개의 데이터 날려버리고 인덱스도 reset
```

```python
pd.concat([df1,df2_right], axis =1)

# concat은 index로 합치기 때문에 안맞네
```

# pd.merge(left, right, on = "기준컬럼", how='--')
- left, right : 이어붙일(병합할) 두 DataFrame
- on : 병합의 기준이 되는 컬럼의 이름
- how : left, right, inner, outer 4가지 병합방식 설정

```python
df2_right
```

```python
pd.merge(
    df1,
    df2_right,
    on="이름",
    how='left'
)

# left기준. 왼쪽에있는거 그대로다둬⭐. 오른쪽에있는거에 이름없어? NaN으로 채워
```

```python
pd.merge(
    df1,
    df2_right,
    on="이름",
    how='right'
)

# right기준? 오른쪽기준⭐. 왼쪽에있어도 오른쪽에 없으면 날려버려
```

```python
df1_left = df1.drop([1,2,6,8])
df1_left
```

```python
pd.merge(
    df1_left,
    df2_right,
    on="이름",
    how='inner'
)
# inner 교집합 양쪽에 모두 있는것만!!
```

```python
pd.merge(
    df1_left,
    df2_right,
    on="이름",
    how='outer'
)
# outer 합집합 양쪽에 있는거 전부다 합쳐
```
