# apply
series, DataFrame에 구체적인 프로그래밍 로직을 적용하고 싶을 때 활용
- 단순 집계 이상으로 프로그래밍적인 효과
  - 데이터에대한 조건검사
  - 형태소 분리
  - 복잡한 조건에 의한 데이터 편집
  - 기타 등등..
- 데이터에 대한 함수를 만들고 싶을 때 aplly를 활용
- apply는 정의한 로직함수를 인자로 넘긴다.

- 반드시 parameter와 return이 있는 함수만 사용가능

```python
# 브랜드 평판지수를 100만으로 나누고, 정수로 캐스팅하고, 짝수인지 홀수인지 알아보는 함수
def brand_even_odd(brand_score):
  brand_score = brand_score / 1000000
  brand_score = int(brand_score)

  return brand_score % 2 == 0

  # true면 짝수 false면 홀수
```

```python
df["브랜드평판지수"].apply(brand_even_odd)
```

```python
# 키 나누기 2
df["키"].apply(lambda x : x/2)

# 너무간단하니 함수만들지말고 람다활용!
```
