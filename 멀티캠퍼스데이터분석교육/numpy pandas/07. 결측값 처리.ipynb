# 결측값 채우기 ⭐⭐
- fillna : NaN값을 특정 값으로 채워주는 함수

```python
df["키"]
```

```python
df["키"].fillna(0)
```

```python
df

# 채운 모양만 보여주는거지 원본을 채워주는건 아님⭐⭐
```

```python
df2 = df.copy()

# df2의 키의 NaN값을 0 으로 채우고 df2에 적용
df2["키"] = df2["키"].fillna(0)  # 실제로 집어넣어줌 키에 < 빈칸을 0으로채운 키를
df2.info()
```

```python
df3 = df.copy()
df3.info()
```

```python
# NaN을 제외한 평균

mean_height = df["키"].mean()
mean_height
```

```python
df3["키"] = df3["키"].fillna(mean_height)
df3

# 키의 NaN자리에 평균값들이 채워졌음
```

```python
# 문제 여자빈칸에 전체평균말고 !! 여자빈칸은 여자끼리 평균으로 채워보기

df4 = df.copy()
```

```python
girl_mean_height = df4.loc[df4["성별"] == "여자", "키"].mean()
girl_mean_height
```

```python
change = df4.loc[df4["성별"] == "여자", "키"].fillna(girl_mean_height)
df4.loc[df4["성별"]=="여자","키"] = change

# 여자 3칸을 < change = NAN값 162.1로 채워준 3칸으로 바꿔줘!
```

```python
df4
```

# 결측치 데이터 제거
- dropna

NaN들어있는거 다삭제

```python
df.dropna()
```

dropna axis활용하기

axis = 0 행삭제

axis = 1 열삭제

```python
df.dropna(axis = 0)
```

```python
df.dropna(axis=1)  # 열삭제
```

## dropna의 how옵션

how = 'any'  : nan값이 1개라도 있으면 제거(기본)

how = 'all'  : 행 또는 열의 데이터가 모두 nan인 경우에만 제거

```python
df.dropna(axis=0, how = 'all')
# 행중에 모두('all') nan값인 행이 없으니까 하나도안날라감
```

```python
import numpy as np
# 행중에 하나를 다 nan값으로 채워보자

df5 = df.copy()
df5.loc[10] = np.nan
df5
```

```python
df5.dropna(axis = 0, how = 'all')
```
