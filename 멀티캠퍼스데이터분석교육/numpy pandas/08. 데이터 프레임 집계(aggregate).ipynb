```python
pd.pivot_table(
    df,  # ëŒ€ìƒì´ ë˜ëŠ” ë°ì´í„° í”„ë ˆì„
    index = "ì†Œì†ì‚¬",  # í–‰ ì¸ë±ìŠ¤ë¡œ ì‚¬ìš©í•  ì§€ì •
    columns = ["ì„±ë³„"],   # ì§‘ê³„ëŒ€ìƒì„ ì—´ê¸°ì¤€ìœ¼ë¡œ ë” ì„¸ì„¸íˆ êµ¬ë¶„  []ë¦¬ìŠ¤íŠ¸ì•ˆì— ì—¬ëŸ¬ê°œê°€ëŠ¥
    values=["í‚¤"]   # valuesìƒëµí•˜ë©´ í‚¤, ë¸Œëœë“œí‰íŒì§€ìˆ˜ ë‹¤ë‚˜ì˜´
)
```
