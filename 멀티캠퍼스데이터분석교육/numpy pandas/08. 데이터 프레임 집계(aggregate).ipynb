```python
import pandas as pd

df = pd.read_csv("korean-idol.csv")
df
```

# 피벗 테이블

1. grouping해서 행의 ***index***정함 (ex 소속사별로 묶겠다.)

2. 집계 대상 구분(***columns***)  키를 여자키 남자키 또나눌수있음.

3. 집계대상값***values*** (키 or 브랜드평판지수)결정 / count빼고 나머지통계는 숫자자료밖에 안됨


```python
pd.pivot_table(
    df,  # 대상이 되는 데이터 프레임
    index = "소속사",  # 행 인덱스로 사용할 지정
    columns = ["성별"],   # 집계대상을 열기준으로 더 세세히 구분  []리스트안에 여러개가능
    values=["키"]   # values생략하면 키, 브랜드평판지수 다나옴
)
```

```python
# 집계 함수 지정
import numpy as np

pd.pivot_table(
    df,
    index = "그룹",
    columns="혈액형",
    values="브랜드평판지수",
    aggfunc=[np.sum, np.mean]  # 함수 이름이 콜백으로 넘어간다.
)
```

```python
# 혈액형별, 그룹마다의 브랜드평판지수 합과 평균을 구하세요

pd.pivot_table(
    df,
    index = "혈액형",
    columns="그룹",
    values="브랜드평판지수",
    aggfunc=[np.sum, np.mean]
)
```

# grouping 그루핑⭐⭐⭐⭐⭐⭐
- 데이터를 집계해서 가공하는데 목적
- 데이터를 그룹으로 묶어서 분석할 때 유용

```python
# 소속사별 집계를 할거다
df.groupby("소속사")
```

```python
df.groupby("소속사").count()

# count는 data type에 영향안받고 다나옴.
# 평균, 분산, 표준편차, 중간값이런건 숫자만 집계가능!!
```

```python
df.groupby("소속사").mean()
```

```python
# 혈액형 별 키의 평균
df.groupby("혈액형").mean()["키"]

# 이건 혈액형별표 다만들고 키columns만 가져온거. 필요없는 추가 계산했음
```

```python
df.groupby("혈액형")["키"].mean()

# 키만가져다가 평균계산하기!! 성능차이 👍
```

```python
# grouping의 기준을 여러 개
df.groupby(["그룹","혈액형"])["키"].std()

# (8,) series말고
```

```python
df.groupby(["그룹","혈액형"])[["키"]].std()

# (8,1)로 보자! dataframe으로 보자👍⭐
# 키를 fancy indexing 형태로 넣어주기 [[]]
```
