## loc 선택 ⭐⭐⭐⭐⭐⭐
- location의 약자
- loc[행선택, 열선택]
- 행선택 : 행의 인덱스 활용 or True/False로 이루어진 조건색인 💕💕
- 열선택 : 컬럼의 이름(들)

## 값 인덱스❤❤❤❤ = loc

```python
df.loc[ : ,"이름"]  # 행은 모두다 선택하겠다. :
```

```python
df.loc[ : , ["이름", "생년월일"]]
```

```python
df.loc[3:8, ["이름","키"]]  # 원래 end-1인데 걍3~8 번까지나옴 값에대한 슬라이싱이라그래!
```

```python
df_sorted = df.sort_values(by ="키")  # by까먹지말기!
df_sorted
```

```python
df_sorted.loc[3:8, ["이름", "생년월일"]]

# 배열인덱스가아니라 값인덱스로나옴!!⭐⭐⭐
```

```python
df_sorted.loc[13:2, ["이름", "생년월일"]]

# loc는 값인덱스
```

# 조건색인(Boolean indexing) ⭐⭐⭐⭐⭐⭐💕💕💕
- 특정한 조건을 이용해서 그 조건을 만족하는 (True) 데이터를 추출

1. 조건 마스크 생성 : True와 False로 이루어진 배열을 생성
2. 생성된 조건 마스크를 색인 위치에 넣어주면 끝!

```python
# 키가 180보다 크다 조건마스크
df["키"] > 180
```

# 위의 트루펄스 결과를 조건마스크라고함!

```python
df[df["키"] > 180]
```

```python
mask = df["소속사"] == "빅히트"
```

```python
df[mask]
```

# 조건인덱싱, 컬럼선택을 동시에
## - 조건인덱싱은 기본적으로 행을 선택하기 위함

```python
# 키가 180보다 큰사람의 이름
mask = df["키"] > 180
df[mask]["이름"]
```

```python
df[mask][["이름","그룹","키"]]
```

```python
df.loc[mask,["이름", "그룹", "키"]]

# loc의 행부분엔 값인덱스 or 조건색인이 들어간다고했음! 위에서
```

# isin을 활용한 인덱싱

```python
# 소속사 리스트(조건을 위한 리스트)
company_lst = ["빅히트", "플레디스"]

df["소속사"].isin(company_lst)
# 소속사가 빅히트또는 플레디스면 true
```

```python
mask = df["소속사"].isin(company_lst)
# 조건마스크. TF로 이루어진 배열

df.loc[mask]
```

```python
# 그룹이 NaN인 데이터를 확인
df.loc[df["그룹"].isnull()]
```

```python
df.loc[df["키"].isnull()]
```

```python
df["키"].notnull()

# non값이 아닌것들 notnull()
```

```python
df[df["키"].notnull()]
```

```python
df.loc[df["키"].notnull(), ["이름","소속사"]]
```

```python
# 소속사가 플레디스 또는 빅히트인 사람들중에 키가 175 이하인 사람들의 이름, 혈액형, 그룹을 키내림차순으로 정렬

mask1 = df["소속사"].isin(["플레디스", "빅히트"])
mask2 =  df["키"] <= 175

# 조건마스크 먼저 만들어놓고하면 편함!

df1 = df.loc[mask1]
df2 = df1.loc[mask2]
df3 = df2.sort_values(by = "키", ascending= False)

# 정렬 먼저하고 뽑으면됨!

df4 = df3[["이름","혈액형","그룹"]]

df4
```
