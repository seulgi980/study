# 배열의 종류
1. 스칼라
  - 0차원 데이터(데이터의 진행방향의 개수)
  - 측정한 하나의 값.
  ex) 키 = 180.1 나이 = 34 몸무게 = 80.5

2. 벡터
  - 1차원 배열
  - 스칼라가 연속적으로 모여있는 것
  - 3명의 키를 하나로 모은다
  ex) 사람키 = [180.1 173.5, 177.8] 3-dimensional vector(feature)

3. 행렬
  - 2차원 배열 ()
  - 1차원 배열인 벡터가 여러개 모여있는 것.

4. 텐서
  - 3차원 이상의 배열 (()행렬이 뒤로 쭈우욱있는모양)
  - 0 rank tensor = 스칼라
  - 1 rank tensor = 벡터
  - 2 rank tensor = 행렬
  
  텐서로 모든 배열의 모양을 이야기 할 수 있다.

# axis(축)
데이터의 진행 방향이라고 보면됨

축의 index⭐⭐⭐

1차원배열 = 벡터의 축은 한방향. axis = 0번

2차원배열 = 행렬의 축은 행,렬 두방향 axis= 0,1번

  행,렬 순서 ↓0번 →1번


```python
arr = np.arange(5)
arr  # 1차원배열 = 벡터
# → aixs 0번 하나밖에 없음
```

```python
np.max(arr, axis=0)
# 축에서 제일 큰값 찾아줌!
```

# 차원수 확장하기!
  - 보통 딥러닝에서 이미지 분석을 하기 위해 CNN 알고리즘을 화용한다.
  - CNN에 필요한 데이터의 형상이 따로 존재하는데, 이 알고리즘을 활용하기 위한 형상(shape)이 존재한다.
  -(N,H,W,C)의 형식을 유지해야 한다.

예시) 흑백 이미지를 수집했는데, 이 이미지의 형상정보(shape)을 확인해보니 (28,28)이라면?

CNN알고리즘에 이미지를 집어넣기 위해선
(1,28,28,1)의 형태로 만들어 줘야 한다.

## ⭐⭐⭐ np.newaxis

```python
# 흑백 이미지 한장을 가지고 있다!
image = np.random.rand(28,28)
image.shape # 2차원 행렬데이터면 안돼!!
            # 4차원으로 확장시켜주기위해 차원수 확장이필요 (1,28,28,1)로 만들어줘야
```

```python
image_expanded_dim = image[np.newaxis, :, :,np.newaxis] # [newaxis, ... , newaxis]해줘도됨
image_expanded_dim.shape
```

# 차원 쥐어짜기 np.squeeze
  - 개수가 1인 차원의 내용을 없애줌 5X1에서 1같은부분

  (1,28,28,1)에서 1같은부분

```python
temp_squeeze = np.squeeze(image_expanded_dim)
temp_squeeze.shape

# 의미가 있는 정보부분만 남겨줌
```


# 배열의 형상바꾸기 ⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐
  - np.newaxis를 활용하면 축하나추가(차원수를 늘린다)
  - np.squeeze를 활용하면 개수가 1인 차원의 데이터를 삭제(차원수를 줄인다.)

평탄화: 2차원 이상의 배열을 1차원 배열로 평평하게 펴주는것⭐⭐⭐
  - ravel :  원본 배열을 평탄화 시킨 참조배열을 만들어 낸다.(이를 View 또는 Reference라고 한다.)
  - flatten : 원본 배열을 평탄화 시킨 복사된배열을 만들어 낸다.

ravel은 원본 배열에 수정이 가해질 가능성이 있고, flatten은 원본 배열에 영향을 미치지 않는다.

```python
x = np.random.randint(15, size=(3,5))  # 0~15까지수 랜덤
x.shape
```

```python
x
```

```python
temp = np.ravel(x)  # 1차원데이터 열벡터로 펴줘!!
temp.shape 
```

```python
temp
```

```python
temp[0] = 100
temp
```

```python
x 
# lst1 = lst2 ⭐참조🌜배열이라 temp바꾸면 원본x도 같이바뀜!
```

```python
x[1, 3] = 9999
temp

# x 바꿔도 temp도 같이바뀜
```

```python
print(hex(id(x)))  # 원본
print(hex(id(temp)))  # 거울

print(hex(id(temp.base))) # 원본

# 근데 x랑 temp메모리 주소 다름.
# temp.base가 같음
# view또는 reference라고함
```

# 브로드캐스팅
  - 차원 수가 다른 배열끼리 연산!
  - 차원이 달라서 0차원 스칼라의 개수가 똑같으면 연산가능
  - ⭐저차원 배열을 >> 고차원 배열⭐|로 확장
  작은걸 올려줘야 큰거만지면안됨
  - 수학적으로 따지게 되면, 고차원 배열의 shape과 똑같은 1벡터(ones)를 만들어서 저차원 배열과 곱한 배열을 만들어서 계산

```python
# shape이 다를 때의 연산
a = np.arange(12).reshape(4, 3) # (4, 3)
b = np.arange(100, 103)         # (3, ) 원소가 3개인 1차원배열 배열은 가로로보기 ㅁㅁㅁ⭐⭐⭐
c = np.arange(1000, 1004)       # (4, )
d = b.reshape(1, 3)             # (1, 3) 행벡터
e = np.arange(0, 4).reshape(4, 1) # ( 4, 1) 열벡터
```

```python
a + b
```

```python
# a + c는 안됨 가로길이가 안맞앙
a + d  # (3,) 도되고 (1,3)도 됨!
```

```python
a + e
```
