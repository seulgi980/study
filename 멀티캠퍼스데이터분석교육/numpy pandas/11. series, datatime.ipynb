# 날짜 변환하기(datetime타입)⭐⭐⭐⭐⭐⭐
- # to_datetime

```python
df["생년월일"]
```

```python
df["생년월일"] = pd.to_datetime(df["생년월일"]) #⭐⭐
df["생년월일"]
```

## to_datetime()했을 때 이점
- 년, 월, 일, 요일, 등등 날짜정보를 바로추출가능
- datetime의 약어인 dt에서 다양한 정보 추출 가능

```python
df["생년월일"].dt.dayofweek  #요일

# 월요일이 0 일요일이 6⭐
```

```python
df["생년월일"].dt.year # 연도
```

```python
df["생년월일"].dt.month # 달
```

```python
df["생년월일"].dt.day # 일
```

```python
df["생년월일"].dt.isocalendar().week  # 현재 날짜가 몇주차인가
```

```python
# dt.minute dt.second, dt.hour도있음
```

```python
# 태어난 연도별 연봉 평균
df['year'] = df['생년월일'].dt.year  # 년도 열추가
df['month'] = df['생년월일'].dt.month  # 월 열추가

df.head()
```

```python

total_df = pd.merge(
    df,
    df2,
    on='이름'
)

total_df.head()
```

```python
total_df.groupby("year")[['연봉']].mean()
```

```python
# 태어난 월별 혈액형 카운트
df.groupby(["month", "혈액형"]).count()
```

```python
df.groupby(["month", "혈액형"])[["혈액형"]].count()
```

```python
# 태어난 월별 혈액형 카운트
import numpy as np

pd.pivot_table(
    total_df,
    index = 'month',
    columns = "혈액형",
    values = "키",  # 암거나넣기 하나만 보기위해
    aggfunc = np.size
)
```
