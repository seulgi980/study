## CSV 파일 읽어오기
**CSV 파일이란??**
- comma sperated value. 데이터가 쉼표로 구분된 거
- 엑셀로 열수도 있지만, 엑셀은 데이터와 관련이 없는 다양한 정보도 포함(ex 표의 굵기, 색상, 계산식...)
- csv를 쓰면 순수하게 데이터만 가져올 수 있고 속도도 빠르다!

```python
pd.read_csv("korean-idol.csv")
```

TIp) 가끔 csv가져오면 깨지는 경우가있음 해결법.
