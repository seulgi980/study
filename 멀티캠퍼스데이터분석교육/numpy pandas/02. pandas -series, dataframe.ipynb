# semi structured data : 준 정형데이터
- Dtype이 없음.거의 문자형으로 통일. 항목개념만 존재
- 스키마가 없음(항목을 컬럼으로 보긴 쫌 애매) filed or property라는 개념을 더 많이 사용.

-ex ) JSON, XML, Parquet

# un structured data : 비정형 데이터
- 형식이 없는 데이터
- 데이터마다 크기도 제각각

- ex) 이미지, 자연어

# Dataframe 만들기
- 시리즈가 여러개 모여서 하나의 데이터 시트를 표현하면 dataframe이 된다.

## 방법1. list로만 만들기
- 컬럼의 이름을 따로 지정해야 할 상황이 생김

```python
sample_data = [['AAA', 34, 180.1],
               ['BBB', 33, 175.5],
               ['CCC', 28, 183.2]]

pd.DataFrame(sample_data)
```

```python
pd.DataFrame(sample_data, columns=["이름", "나이", "키"])
```

## 방법2. dict로 만들기
- key값은 컬럼의 이름이된다.
- ⭐⭐⭐value는 반드시 리스트로 이루어져 있어야 한다. (시리즈)
- 모든 value의 리스트는 반드시 길이가 맞아야 한다.



```python
sample_data = {
    "이름" : ["AAA","BBB","CCC"],
    "나이" : [34, 34, 28],
    "키" : [180.1, 175.7, 183.2]

}

pd.DataFrame(sample_data)
```

## dict로 만들기(JSON)
- json= javascript object natation

```python
sample_data = [
  {
    "이름" : "AAA",
    "나이" : 34,
    "키": 180.1
  },
  {
    "이름": "BBB",
    "나이" : 33,
    "키": 175.5
  },  
  {
    "이름" : "CCC",
    "나이" : 28,
    "키": 183.2
  }
]

sample_data

# 리스트안에 딕셔너리들이 들어있음
```
