# 데이터의 추가와 삭제

## row 추가하기
- dict형식의 데이터를 만들어 준 다음 append() 함수를 활용해 데이터를 추가

- 반드시 ***ignore_index = True***  안하면 에러!

```python
new_data = {
    "이름": "A",
    "그룹": "뉴 그룹",
    "소속사": "돈 많이 주는 소속사",
    "성별": "남자",
    "생년월일": "1988-01-29",
    "키": 180.1,
    "혈액형": "AB",
    "브랜드평판지수": 123456789
}
```

```python
df.append(new_data, ignore_index = True)  # 반드시반드시 ignore_index = True로 해줘야!!⭐⭐

# lst append 했을땐 결과 암것도 안나왔는뎅 이건 나왔네?
```

```python
df.tail()

# 원본 바꿔준게아님!! 추가하면 저런 결과가 나올거야 = 복사본생성한거⭐⭐⭐
```
