# 데이터 정렬

```python
## 오름차순 인덱스 정렬
```

```python
df.sort_index() # 오름차순정렬이 기본값임. ascending = True
```

```python
# 내림차순 정렬
df.sort_index(ascending = False)
```

# 컬럼 오름차순 정렬
- by 옵션을 이용해서 정렬의 기준이 되는 컬럼을 지정. 2개 이상도 가능
- ascending 옵션을 이용해서 오름/내림차순 설정가능

```python
# 키 오름차순으로 정렬
df.sort_values(by = "키")
```

```python
df.sort_values(by = "키", ascending = False)
```

### 2개 이상 컬럼 정렬하기!
- 앞쪽컬럼이 먼저 정렬됨⭐⭐⭐
- 키로 먼저 정렬 쭉하고 뒤에꺼 정렬됨

by = ["키", "브랜드평판지수"]

```python
df.sort_values(by = ["키", "브랜드평판지수"])
```

```python
df.sort_values(by = ["키", "브랜드평판지수"], ascending = [False, True])

# 키 내림차순/ 브랜드평판 오름차순 으로 정렬해라!
```

```python
df.sort_values(by="이름")
# 문자열데이터로도 정렬 가능!
```
