# 중복 데이터 제거

```python
df["그룹"]
```

```python
df["그룹"].drop_duplicates()

# drop_duplicates() 중복제거!
# kepp = 'first'옵션 위에서부터 첫 방탄소년단 찾아서 그 인덱스
```

keep 옵션을 사용해서 유지하고 싶은 데이터를 선택할 수 있다.

- keep = 'first' : 첫번째 데이터를 남긴다(기본)
- keep = 'last' : 제일 마지막 데이터를 남긴다.

```python
df["그룹"].drop_duplicates(keep = 'last')

# 아래서부터 첫 방탄소년단 찾아서 그인덱스
# 아래서부터 첫 뉴이스트 찾아서 그인덱스
```

# 행/열 삭제 drop

```python
df.drop("그룹",axis = 1)  # axis 기본값은 0임
```
