# 파싱 (parsing)⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐증말증말 중요

문자열을 >> 다른자료형으로 바꾸는 과정

```python
score1 = "9.49"
score2 = "9.75"
# 웹에서 가져온 데이터는 100% 문자열이기때문에 영화평점 더하고싶다.
# 이런경우는 (문자를 > 실수화)시켜서 평균구해야
```

```python
# parsing
score1 = float(score1)
score2 = float(score2)

score1 + score2
```

```python
score1 = "9.15"
int(score1) #오류남 파싱 오류 9.15는 정수가 아니라 정수로 변환할 수 없기 때문
```

```python
score2 = "8"
float(score2) #이건됨 ㅋ ㅋ
```

```python
score3 = "9.15점"
float(score3) # 점이란 글자때문에 안됨
```
