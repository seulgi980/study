# 멤버 연산자
- 멤버란? 컬렉션 안에 원소들 item
컬렉션에 원소가 존재하는지 여부를 판단할때!!'
in   not in

```python
lst = ["a", 1, 3.1]
"a" in lst
```

```python
"b" in lst
```

```python
"b" not in lst
```

```python
tem_dict ={
    "a" : 1,
    "b" : 2
}
```

```python
#기본적인 멤버연산은 ⭐key의 존재여부를 물엉
"a" in tem_dict
```

```python
1 in tem_dict
```

```python
#그럼 1 이 value에 들어있는지 확인할려묜?
1 in tem_dict.values()
```
