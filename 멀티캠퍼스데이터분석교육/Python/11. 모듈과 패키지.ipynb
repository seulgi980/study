# 모듈 module, 패키지 package

- 모듈 : 변수, 함수, 클래스를 모아놓은 .py 확장자 파일
  - .py : 마크다운, 셀같은거 없음 순수한 파이썬 코드만 존재
  - .ipynb : 마크다운 및 시각화 등 여러 정보를 예쁘게 볼 수 있도록 꾸며진 파일. (코랩 노트북같은거)

-패키지 : 모듈의 기능을 디렉토리(폴더)별로 정리해 놓은 개념

```python
%%writefile demo.py

num = 10

def foo(msg):
  print(f"foo() => {msg}")

def goo(num1, num2):
  # 여러 개의 리턴 값을 튜플로 묶어서 리턴
  return num1+num2, num1-num2, num1*num2, num1/num2

class DemoCalc:
  def sum_all(self, *args):
    return sum(args)
```

```python
foo("Hello") # 오류 모듈 불러와야함!
```

# 모듈 불러오기
import 모듈의 이름

```python
import demo
# .py파일 이름이 모듈의 이름
```

```python
demo.num
```

```python
demo.foo("Hello")
```

```python
demo.goo(10,20)
# 하나의 튜플로 묶어서 여러개 리턴값줌.
# 언패킹도 가능하겠죠?
```

```python
item1, item2, item3, item4 = demo.goo(10,20)

print(item2)
```

```python
calc = demo.DemoCalc()
calc.sum_all(1,2,3,4,5,6,7,8,9)
```

모듈 import할 때 별명 지어줄 수 있음!!
alias⭐⭐

- 모듈의 이름이 너무 길어서 불편하다거나 그럴 때

```python
import demo as dm
```

```python
dm.foo("하이하이")  #별명으로 사용가능!
```

모듈 내에 있는 함수, 변수, 클래스를 따로따로 불러오기 가능!

-from 모듈이름 import 변수명, 함수명, 클래스명

```python
from demo import num, goo
# from을 써서 모듈안의 몇개만 불러오기
```

```python
print(num)
```

```python
goo(10,20)  # 모듈 이름 없이 바로사용 가능!!
```

# 패키지 (package)
! : 리눅스 명령어 (운영체제)

```python
!mkdir -p mc_pkg/pkg1
!mkdir -p mc_pkg/pkg2
```

```python
%%writefile mc_pkg/pkg1/my_module.py

def my_func(msg):
  print("Here is pkg1.my_module")
  print("Good bye")
```

```python
%%writefile mc_pkg/pkg2/your_module.py

def your_func():
  print("여기는 pkg2의 your_module 입니다.")
  print("Hi~")
```

```python
%%writefile mc_pkg/pkg2/hello.py

def hello_func():
  print("Hello~")
```

```python
# 이제 패키지를 거쳐서 모듈을 불러와야댐
import mc_pkg.pkg2.hello  # 맨 마지막엔 파일의 이름이 나와야대!!
```

```python
mc_pkg.pkg2.hello.hello_func()
```

```python
import mc_pkg.pkg2.hello as hi
# 마찬가지로 너무 기니까 alias 사용해서 별명으로 부르기가능!

hi.hello_func()
```

```python
from mc_pkg.pkg1.my_module import my_func

my_func("하이하이")
```
