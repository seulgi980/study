#decorator
코드를 바꾸지 않고 기능을 추가하거나 수정하고 싶을  때 사용하는 문법

```python
def tag_card(callback):

  def wrapper(*args, **kwargs):
    print("카드를 찍고 탑니다.")
    callback(*args, **kwargs)  #튜플에 *을 붙이면 튜플의 내용이 다 풀려서 들어간다.
                               #args해서 넘기면 (1,2,3)   *args 해서 넘기면 1,2,3,으로 넘어간다.
    print("카드를 찍고 내립니다.")

  return wrapper  #내가 원할때 실행시켜줄 수 있다!! (버스가 실행될 때)
```

```python
#decorator
@tag_card
def bus(fee):
  print(f"{fee}를 내고 버스를 탑니다")

bus(1250) #오오 버스가 메인으로 돌아왔덩👍👍
```

```python
def subway(fee):
  print(f"{fee}를 내고 지하철을 탑니다.")

wrapper = tag_card(subway) #subway를 집어넣은 wrapper가 된거임 내가 원할 때 불러오면됨
wrapper(1250)
```

```python
# 여기에 새로운 함수를 추가하고싶으면
# def 새함수() print~~ 다시 다쳐야한다. 나는 틀은 그대로고 함수이름만 바꾸고싶다.
# 데코레이터는 함수를 수정하지 않은 상태에서 추가 기능을 구현할 때 사용합니다. 지금 딱 필요함!
```

```python
def trace(func):                             # 호출할 함수를 매개변수로 받음
    def wrapper():                           # 호출할 함수를 감싸는 함수
        print(func.__name__, '함수 시작')    # __name__으로 함수 이름 출력
        func()                               # 매개변수로 받은 함수를 호출
        print(func.__name__, '함수 끝')
    return wrapper                           # wrapper 함수 반환

# 이렇게 trace 데코레이터만 하나 추가해주면 간단하게 할 수 있다!

def hello():
    print('hello')

def world():
    print('world')

trace_hello = trace(hello)    # 데코레이터에 호출할 함수를 넣음
trace_hello()                 # 반환된 함수를 호출
trace_world = trace(world)    # 데코레이터에 호출할 함수를 넣음
trace_world()                 # 반환된 함수를 호출
```

```python
def trace(func):                             # 호출할 함수를 매개변수로 받음
    def wrapper():
        print(func.__name__, '함수 시작')    # __name__으로 함수 이름 출력
        func()                               # 매개변수로 받은 함수를 호출
        print(func.__name__, '함수 끝')
    return wrapper                           # wrapper 함수 반환

@trace    # @데코레이터 위의 trace_hello = trace(hello)안쓰고 편하게 한 것
def hello():
    print('hello')

@trace    # @데코레이터
def world():
    print('world')

hello()    # 함수를 그대로 호출
world()    # 함수를 그대로 호출
```

```python
@trace    # @데코레이터
def hello():
    print('hello')

@trace    # @데코레이터
def world():
    print('world')

hello()    # 함수를 그대로 호출
world()    # 함수를 그대로 호출
```
