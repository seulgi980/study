#í´ë¡œì € (closure)
ë°– ì˜ì—­ì—ì„œ ì§€ì—­ì˜ì—­ì— ì ‘ê·¼í•˜ëŠ” ê²ƒì´ ì›ì¹™ì ìœ¼ë¡  ì•ˆë˜ëŠ”ë°

```python
#outer ì•ˆì— ìˆëŠ” innerë¥¼ outerì˜ í˜¸ì¶œ ì‹œì ì— ê°™ì´ ì‹¤í–‰í•˜ëŠ”ê²Œ ì•„ë‹ˆê³ , ì›í•˜ëŠ” ì‹œì ì— ì‹¤í–‰í•˜ëŠ” ë°©ë²•
def outer(a,b):
  print("Here is Outer")

  def inner(c,d):
    print("Here is Inner")
    print(a,b,c,d,)
    return a+b+c+d

  return inner  #í•¨ìˆ˜ ìì²´ë¥¼ ë¦¬í„´
```

```python
inner_func = outer(10,20)
inner_func
#í•¨ìˆ˜ ìì²´ë¥¼ ë¦¬í„´
#10, 20ì„ ë¯¸ë¦¬ ë°›ì•„ë†¨ë‹¤ê°€
```

```python
def outer():
  o_a = 10
  o_b = 20

  print("Here is Outer", o_a, o_b)

  def inner(a,b):
    o_a = a
    o_b = b

    print("Here is Inner", o_a, o_b)

  return inner
```

```python
def tag_card(callback):

  def wrapper(*args, **kwargs):
    print("ì¹´ë“œë¥¼ ì°ê³  íƒ‘ë‹ˆë‹¤.")
    callback(*args, **kwargs)  #íŠœí”Œì— *ì„ ë¶™ì´ë©´ íŠœí”Œì˜ ë‚´ìš©ì´ ë‹¤ í’€ë ¤ì„œ ë“¤ì–´ê°„ë‹¤.
                               #argsí•´ì„œ ë„˜ê¸°ë©´ (1,2,3)   *args í•´ì„œ ë„˜ê¸°ë©´ 1,2,3,ìœ¼ë¡œ ë„˜ì–´ê°„ë‹¤.
    print("ì¹´ë“œë¥¼ ì°ê³  ë‚´ë¦½ë‹ˆë‹¤.")

  return wrapper  #ë‚´ê°€ ì›í• ë•Œ ì‹¤í–‰ì‹œì¼œì¤„ ìˆ˜ ìˆë‹¤!! (ë²„ìŠ¤ê°€ ì‹¤í–‰ë  ë•Œ)
```

```python
#decorator
@tag_card
def bus(fee):
  print(f"{fee}ë¥¼ ë‚´ê³  ë²„ìŠ¤ë¥¼ íƒ‘ë‹ˆë‹¤")

bus(1250) #ì˜¤ì˜¤ ë²„ìŠ¤ê°€ ë©”ì¸ìœ¼ë¡œ ëŒì•„ì™”ë©ğŸ‘ğŸ‘
```

```python
def subway(fee):
  print(f"{fee}ë¥¼ ë‚´ê³  ì§€í•˜ì² ì„ íƒ‘ë‹ˆë‹¤.")

wrapper = tag_card(subway) #subwayë¥¼ ì§‘ì–´ë„£ì€ wrapperê°€ ëœê±°ì„ ë‚´ê°€ ì›í•  ë•Œ ë¶ˆëŸ¬ì˜¤ë©´ë¨
wrapper(1250)
```

```python
# ì—¬ê¸°ì— ìƒˆë¡œìš´ í•¨ìˆ˜ë¥¼ ì¶”ê°€í•˜ê³ ì‹¶ìœ¼ë©´
# def ìƒˆí•¨ìˆ˜() print~~ ë‹¤ì‹œ ë‹¤ì³ì•¼í•œë‹¤. ë‚˜ëŠ” í‹€ì€ ê·¸ëŒ€ë¡œê³  í•¨ìˆ˜ì´ë¦„ë§Œ ë°”ê¾¸ê³ ì‹¶ë‹¤.
# ë°ì½”ë ˆì´í„°ëŠ” í•¨ìˆ˜ë¥¼ ìˆ˜ì •í•˜ì§€ ì•Šì€ ìƒíƒœì—ì„œ ì¶”ê°€ ê¸°ëŠ¥ì„ êµ¬í˜„í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤. ì§€ê¸ˆ ë”± í•„ìš”í•¨!
```

```python
def trace(func):                             # í˜¸ì¶œí•  í•¨ìˆ˜ë¥¼ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ìŒ
    def wrapper():                           # í˜¸ì¶œí•  í•¨ìˆ˜ë¥¼ ê°ì‹¸ëŠ” í•¨ìˆ˜
        print(func.__name__, 'í•¨ìˆ˜ ì‹œì‘')    # __name__ìœ¼ë¡œ í•¨ìˆ˜ ì´ë¦„ ì¶œë ¥
        func()                               # ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ì€ í•¨ìˆ˜ë¥¼ í˜¸ì¶œ
        print(func.__name__, 'í•¨ìˆ˜ ë')
    return wrapper                           # wrapper í•¨ìˆ˜ ë°˜í™˜

# ì´ë ‡ê²Œ trace ë°ì½”ë ˆì´í„°ë§Œ í•˜ë‚˜ ì¶”ê°€í•´ì£¼ë©´ ê°„ë‹¨í•˜ê²Œ í•  ìˆ˜ ìˆë‹¤!

def hello():
    print('hello')

def world():
    print('world')

trace_hello = trace(hello)    # ë°ì½”ë ˆì´í„°ì— í˜¸ì¶œí•  í•¨ìˆ˜ë¥¼ ë„£ìŒ
trace_hello()                 # ë°˜í™˜ëœ í•¨ìˆ˜ë¥¼ í˜¸ì¶œ
trace_world = trace(world)    # ë°ì½”ë ˆì´í„°ì— í˜¸ì¶œí•  í•¨ìˆ˜ë¥¼ ë„£ìŒ
trace_world()                 # ë°˜í™˜ëœ í•¨ìˆ˜ë¥¼ í˜¸ì¶œ
```

```python
def trace(func):                             # í˜¸ì¶œí•  í•¨ìˆ˜ë¥¼ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ìŒ
    def wrapper():
        print(func.__name__, 'í•¨ìˆ˜ ì‹œì‘')    # __name__ìœ¼ë¡œ í•¨ìˆ˜ ì´ë¦„ ì¶œë ¥
        func()                               # ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ì€ í•¨ìˆ˜ë¥¼ í˜¸ì¶œ
        print(func.__name__, 'í•¨ìˆ˜ ë')
    return wrapper                           # wrapper í•¨ìˆ˜ ë°˜í™˜

@trace    # @ë°ì½”ë ˆì´í„° ìœ„ì˜ trace_hello = trace(hello)ì•ˆì“°ê³  í¸í•˜ê²Œ í•œ ê²ƒ
def hello():
    print('hello')

@trace    # @ë°ì½”ë ˆì´í„°
def world():
    print('world')

hello()    # í•¨ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ í˜¸ì¶œ
world()    # í•¨ìˆ˜ë¥¼ ê·¸ëŒ€ë¡œ í˜¸ì¶œ
```
