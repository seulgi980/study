# 예외는 파이썬 문법 (try-except)를 이용해서 처리가 가능

프로그램 구동 중 Exception이 발생하면 무조건 프로그램이 종료가됨

```python
1 / 0
# 오류메시지 띄워주면되지 프로그램을 꺼버릴 필요까ㅓ진 없자나...ㅠㅠ
```

- try : 내가 할일들(예외가 일어날만한 코드가 들어가는 구간)
- except : try에서 일어난 예외를 처리하는 구간
- finally : try, except 구문이 실행된 후 실행할 코드
- raise


try랑 except는 짝궁

```python
lst = [1,2,3]
try:
  print("Hello")
  print(lst[2])
  print("Bye")
except:
  print("Error 입니다")
```

```python
lst = [1,2,3]
try:
  print("Hello")
  print(lst[3])
  print("Bye")
except:
  print("Error 입니다")

  # 예외가 터지면 그 즉시 코드 멈춤 아래쭉안나옴
```

```python
# Exception 객체 활용
num = 1

try:
  print("나눗셈 수행")
  result = 1 /num
  print("나눗셈 종료")

except Exception as e:
  print("에러 발생")
  print(e)
```

```python
# Exception 활용
num = 0

try:
  print("나눗셈 수행")
  result = 1 /num
  print("나눗셈 종료")

except Exception as e:
  print("에러 발생")
  print(e)
```

- finally : try, except 구문이 실행된 후 실행할 코드

try에서 정상적으로 코드가 실행이 됐든, except에서 예외를 처리 했든 무조건 실행되는 구문(후처리 한다고생각)

try나 except에서 만들어진 자원을 해제하는 역할

```python
num = 0

try:
  print("나눗셈 수행")
  result = 1 /num
  print("나눗셈 종료")

except Exception as e:
  print("에러 발생")
  print(e)

print("Done")

#뭔가 동떨어진 context같으니까
```

```python
num = 0

try:
  print("나눗셈 수행")
  result = 1 /num
  print("나눗셈 종료")

except Exception as e:  #Exception도 객체 클래스임!
  print("에러 발생")
  print(e)

finally:
  print("Done")
```

```python
try:
  print("크롤링수행")

except Exception as e:
  print("에러 부분 체크")

finally:
  print("여태까지 모은거 모두 저장")
```

# raise
- 특정 상황에서 예외를 강제로 일으키는 거

```python
# 예외 (Exception) 커스터마이징 - 상속을 통해서!
class AgeException(Exception):
    def __str__(self):           # 객체를 문자열로 표현하기 위해 오버라이딩
      return "나이 값이 올바르지 않습니다"
```

```python
class A:
  pass
print(A())

# 이렇게 정보 다~~나오는게 __str__이용돼서 그런거
# 모든 class객체들 A같은것들은 A(object)가 상속되어있음 object클래스가 상속되어있음 눈엔보이지않지만
```

```python
def set_age(age):
  if age < 0:
    raise AgeException()

  print(age)
```

```python
set_age(10)
```

```python
set_age(-10)
```
