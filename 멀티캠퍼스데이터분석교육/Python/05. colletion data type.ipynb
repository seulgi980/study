# 참조와 복사

```python
lst5 = [1,2,3]
print(lst5.append(4))
# 바뀌기만 한거라 보여줄게 없다. append는 그냥 원본을 바꾸라는거
```

```python
lst5[:2]
# 원본을 바꾼게 아니라 잘라낸 ⭐복사본을 보여주는거(슬라이싱은 복사본)
```

```python
# pop 맨끝에꺼 보여주면서 삭제(쌓이진않음 소모됨)
lst = [1, 2, 3, "three", "four", [5,6], 7.0, 8.1]
lst
```

```python
item = lst.pop()
print(item, "/", lst)
# 컨트롤 엔터 계속해보면 끝에꺼 하나씩 lst에서 사라져서 뽑혀나옴
```

# 패킹 packing 언패킹 unpaking ⭐⭐⭐⭐⭐⭐⭐⭐⭐⭐
- 패킹 : 데이터 모으기
- 언패킹: 모아낸 데이터 풀어놓기

```python
tup = (1, "two", 3)
# 데이터모으는거 이거자체를 패킹이라함
```

```python
#언패킹
item1, item2, item3 = tup
print(item1)
print(item2)
print(item3)
```

```python
item1, item2 = tup
# 갯수 맞아야 언패킹 가능!
```

```python
item1, item2, item3, item4 = tup
```

```python
tup = ("one", 2, "three")
# 관례상 필요없는 데이터에 대한 변수는 언더바를 사용!!!
item1, item2, _ = tup
```

#dictionary⭐⭐⭐⭐⭐⭐⭐
-{key : value}

  -{key1 : value1, key2 : value2}
  키와 밸류가 쌍으로있어야함 페어형태(pair)
##key의 특징
  - 정수, 문자만 가능⭐
  - set구조로 저장 중복허용X순서X
##value특징
  -list형태로 저장 중복o 모든데이터 저장가능~

```python
my_info = {"name":"한슬기","age":25, "height" : 163}
# key = {name, age, height} value = [한슬기, 25, 163]
my_info
```

```python
my_info = {"name" : "한슬기", "nickname" : "한슬기"}
# value는 리스트로 저장이라 중복가능!
my_info
```

```python
my_info = {"name" : "한슬기", "name" : "바보야"}
my_info
# 둘중 하나만나옴 key는 중복되면 안됨
```

```python
my_info = {
    "name" : "한슬기",
    "age" : 25,
    "height" : 163
}
# 데이터 추가 수정가능 DIct
```

```python
#추가 *대괄호*안에 key값!!
my_info["address"] = "의왕시"
my_info
```

```python
my_info["address"] = "경기도 의왕시"
my_info
# key가 있으면 수정 key가 없는거였으면 추가
```

```python
# 딕셔너리에서 데이터 가져올땐? key를 물어봐!
my_info["height"]
```

```python
my_info["weight"]
#없는 key는 에러
```

```python
#key 가 존재하지 않는 경우 지정한 값을 보여주는 기능
my_info.get("weight","몸무게 정보 없음")
```

```python
#삭제하기 del사용
del my_info["address"]
my_info
```

```python
my_info["adress"] = "공주집"
my_info
```
