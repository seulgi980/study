# 클래스의 구조
- 멤버변수, (멤버)메소드
- 멤버변수 = data, state
- 멤버 메소드 = 기능( 함수는 클래스 밖에서 자체기능 메소드는 객체가 해야 할 기능)

car클래스 만들어보기!
- 데이터(멤버변수) : 브랜드, 가격, 중량, 색상...
- 기능(메소드) : 에어컨켜기, 시동걸기, 브레이크 밟기, 악셀 밟기, 시동 끄기...

```python
#클래스의 이름은 PascalCase로 만들기!!
class MyCar:
  # 멤버 변수 선언 부분
  brand = "벤츠"
  price = 10000

  # 메소드 선언 부분
  def drive(self):  #self가 들어가면 멤버 (메소드)다! 라고 보면됨
                    # self말고 김치 감자라고 적어도 똑같은데 관례상 self로 적어야됨.
    print("차가 앞으로 갑니다")

  def brake(self):
    print("차가 멈춥니다.")

```

# self ⭐⭐⭐⭐⭐
- 객체 자기 자신을 참조하는 변수
- self가 아닌 다른 변수명을 사용할 수 있지만 권장하지 않음.

- 클래스 내에서 ***멤버변수*** ***멤버 메소드*** 에 접근 할 때 사용
***멤버 메소드*** 정의 시 사용


```python
class Person:
  def __init__(self, name, job, age):
    self.name = name
    self.job = job
    self.age = age

  def print_info(self):
    print(hex(id(self))) #self의 메모리 주소값 확인
    print(f"{self.name}의 직업은 {self.job}입니다.")
```

```python
민호 = Person("소민호", "강사", 10)
민호.print_info()
```

```python
print(hex(id(민호)))
#민호랑 self랑 메모리 주소값 같음!!
```

```python
진호 = Person("소진호", "연구원", 10)
진호.print_info()  #self의 메모리 주소값
```

```python
print(hex(id(민호.age)))
print(hex(id(진호.age))) # 민정님 질문 primitive자료형은 하나에저장
```

-부모로부터 받는 메소드를 자식 클래스가 ***수정***하고 싶을 때 (재정의 한다)

# 오버라이딩의 규칙
- 부모클래스의  메소드형식(이름, parameter 등)을 자식 클래스에서 똑같이⭐⭐ 따라해야됨!!

```python
class A:
  def foo(self):
    print("Here is A.foo")

  def goo(self, param):
    print(f"Here is A.goo, {param}")
```

```python
#A클래스의 자식클래스 만들어보장!
class B(A):
  def koo(self):
    print("Here is B.koo")
    self.foo() # A를 상속받아서 B는 A의 기능이 포함되어있다. 그러므로 self(나=B의) foo는 A에서받은 A의foo
```

```python
b = B()
b.koo()
```

```python
b.foo()
```

```python
b.goo("파라미터어")
```

```python
class C(A):
  def moo(self):
    print("Here is C.moo")


  # A클래스의 foo랑 goo메소드를 오버라이딩
  def foo(self):
    print("Here is C.foo")

  def goo(self, param):
    print(f"Here is C.goo, {param}")
    print("ㅎㅇㅎㅇ")
```

```python
c = C()
c.moo()
c.foo()  #자식이 재정의한게 우선적으로 나옴
c.goo("파라미터파라미터")



a= A()
a.foo()
```

# 상속의 법칙

- 자식클래스의 객체를 생성하면 부모클래스의 객체가 ***먼저*** 만들어진다
-즉, 부모클래스의 생성자가 먼저 호출되고 나서, 자식클래스의 생성자가 호출된다.

부모클래스의 생성자에 parameter가 존재한다면 어쩌지..?

```python
class A:

  def __init__(self,msg):
    print("부모클래스 A의 생성자가 호출됩니다.")
    self.msg = msg

  def print_msg(self):
    print(f"{self.msg}")
```

```python
class B(A):

  def foo(self):
    print("Here is B.foo")
```

```python
b = B()  #A클래스에 넣어줄 msg가 없으니까 오류가남!  __init__이라서 초기값 꼭 필요하기때문에!!
# init을 joo같은걸로 바꾸면 오류안남

# 자식 클래스에서는 부모 클래스의 생성자⭐⭐를 책임을 져야한다.
# 자식클래스에서 부모클래스의 parameter를 책임진다.
```

```python
class B(A):

  def __init__(self,msg):
    # 부모클래스의 생성자에 직접적으로 접근해서 argument를 넘겨줘야한다.(올려준다라고 표현super)
    super(B, self).__init__(msg)

  def foo(self):
    print("Here is B.foo")
```

```python
b = B("B에서 보내는 메시지")  #B(자식클래스)에서 받은 msg를 A로 올려준다!!super
# A init에 들어가서 self.msg = "B에서 보내는 메시지"들어감!!
```

```python
b.print_msg()
```

# Getter & Setter
- 멤버 변수에 접근할 때 ***특정 로직을 거쳐서*** 접근시키는 방법

변수에 값 집어넣기 Setter

변수를 가져올 때 Getter

```python
class User1:
#  init은 처음에 값만 ⭐넣어주면되는데 왜 <검사>를 하고있지?... 분리할꺼야

  def __init__(self, name):
    # 글자 수가 3글자 이상이 되어야만 회원이름이 등록이 된다.
    if len(name) >= 3:
      self.name = name
    else:
      self.name = None
      print("길이가 짧습니다.")
```

```python
user1 = User1("a")
user1.name
```

```python
# 멤버 변수를 검사하는 기능이 들어가 있어야 할까?
```

```python
class User2:

  def __init__(self, first_name):
    self.first_name = first_name

  # 멤버 변수 first_name에 "값이 들어갈 때 수행"되는 메소드
  def setter(self, first_name):
    print("Set First Name")

    if len(first_name) >= 3:
      print("Set first name success")
      self.first_name = first_name
    else:
      print("Set first name failed")
      self.first_name = None

  # 멤버 변수 first_name을 "가져 올 때" 수행되는 메소드
  def getter(self):
    print("Get first name UPPER")
    return self.first_name.upper()


  # setter, getter를 등록하기
  name = property(getter, setter)
```

```python
user2 = User2("m")
user2.first_name
# setter안쓰고 그냥 init으로 한거
```

```python
user2.name = "m"   # .name해서 setter쓰기 검사되어서 들어감
```

```python
user2.name = "minho" # .name해서 setter쓰기 검사되어서 들어감
```

```python
user2.name  # getter 변수 가져올 때 수행되는 메소드
```

# 맹글링 (Mangling) 외부로부터 직접적인 접근 차단.

```python
class User3:

  def __init__(self, first_name, age):

    # 맹글링이란?
    # 멤버 변수, 메소드의 이름 앞에 언더바 두번(__) 붙여서 짓는 기법⭐

    self.__first_name = first_name # first_name은 맹글링이 된 상태
    self.age = age # age는 맹글링이 되지 않은 상태

  # 맹글링 된 변수는 클래스 내부에서는 언제든지 사용할 수 있다.
  def getter(self):
    return self.__first_name

  def setter(self, first_name):

    if len(first_name) >= 3:
      print("[Success] Set First name")
      self.__first_name = first_name
    else:
      print("[Error] Set First name")
      self.__first_name = None

  name = property(getter, setter)
```

```python
user3 = User3("minho", 35)
```

```python
user3.name = "m"
```

```python
user3.name = "minho"
```

```python
user3.name
```

```python
user3.__first_name = "m"
# 입력해도 name에는 영향 없음
```

```python
user3.name #안바꼈죵?
```

```python
user3_1 = User3("minho",35)
user3_1.__first_name  #맹글링 해놔서 접근 불가
```

```python
user3.__first_name = "집어ㅏ넣기 name바껴라아ㅏ"
```

```python
user3_1.name #안바뀜
```

# 멤버변수말고 메소드도 맹글링해보자

```python
class User4:

  def __init__(self, first_name, age):

    # setter를 이용해서 초기화
    self.set_first_name(first_name)  #self.first_name = first_name이렇게 해도됨
    self.age = age   #self.set_age(age) 해도됨

  # first_name에 대한 setter
  def set_first_name(self, first_name):

    if len(first_name) >= 3:
      self.__first_name = first_name
    else:
      self.__first_name = None

  # age에 대한 setter
  def set_age(self, age):

    if age > 0 :
      self.__age = age
    else:
      self.__age = None

  # first_name에 대한 getter
  def get_first_name(self):
    # 이름 중간의 모든 글자가 *로 표시 될 수 있도록
    # minho -> m***o
    first_char = self.__first_name[0]
    last_char  = self.__first_name[-1]
    star_len   = len(self.__first_name)-2
    stars      = "*" * star_len

    return f"{first_char}{stars}{last_char}"

  # age에 대한 getter
  def get_age(self):
    return self.__age

  # 사람의 정보를 표기하기 위한 문자열을 만드는 메소드를 구현
  def __make_user_info(self):
    return f"{self.get_first_name()}의 나이는 {self.get_age()}"

  def print_user_info(self):
    user_info_text = self.__make_user_info()
    print("⭐", user_info_text, "😊")

  # getter & setter 등록
  first_name = property(get_first_name, set_first_name)
  age = property(get_age, set_age)
```

```python
user4 = User4("minho", 35)
```

```python
user4.print_user_info()
```

```python
user4.__make_user_info()  #맹글링 해놔서 안됨
```

```python
class Engine:

  def turn_on(self):
    print("엔진 시동을 걸었습니다.")

  def turn_off(self):
    print("엔진 시동을 껐습니다.")
```

```python
class Car(Engine): # 자동차는 엔진입니다. - 성립하지 않는 명제

  def drive(self):
    self.turn_on()
    print("자동차가 앞으로 갑니다.")

  def stop(self):
    print("자동차가 멈췄습니다.")
    self.turn_off()
```

```python
car = Car()
car.drive()
car.stop()
```

## has a 관계
 - A has a B : 클래스 A는 클래스 B를 갖는다.
 B 는 A의 부품이다.

 -A클래스의 멤버변수로 B클래스의 객채를 가지고 있으면 됨!

자동차는 엔진이다 - is a    X


자동차는 엔진을 갖는다. 엔진은 자동차의 부품이다!!
has a   O

```python
# 엔진도 카테고리가 될 수 있음. 람보르기니 엔진, 봉고엔진이 엔진카테고리의 항목들
# Car카테고리의 하위항목들 bongo,sonata, lambo
# 이러케 생각하면 쉬움
class Engine:
  def __init__(self, name):
    self.name = name
  def turn_on(self):
    print(f"{self.name} 엔진 시동을 걸었습니다.")
  def turn_off(self):
    print(f"{self.name} 엔진 시동을 껐습니다.")

class LamboEngine(Engine):
  def __init__(self):
    super(LamboEngine, self).__init__("람보르기니")
```

```python
class Car:

  def __init__(self):
    # 무조건 람보르기니 엔진만 갖다가 사용하는 상태
    self.engine = LamboEngine() # 강결합 상태. 사용할 클래스가 어떤 클래스 인지를 확실하게 알고 있는 상태

  def drive(self):
    # 내 부품 중에 하나인 engine의 시동을 켠다.
    self.engine.turn_on()
    print("자동차가 앞으로 갑니다.")
```

```python
car = Car()
car.drive()
```

```python
# 엔진에 대한 약결합 구현

# 그냥 추상적으로 자동차니까 어떤 엔진이 들어갈지도 추상적으로 설정
class Car:

  def __init__(self, engine):
    self.engine = engine # 어떤 엔진을 Car 클래스에서 부품으로 사용할지 몰라요... -> 약결합

  def drive(self):
    # 내 부품 중에 하나인 engine의 시동을 켠다.
    self.engine.turn_on()
    print("자동차가 앞으로 갑니다.")
```

```python
class BongoEngine(Engine):
  def __init__(self):
    super(BongoEngine, self).__init__("봉고")
```

```python
car = Car(LamboEngine())
car.drive()
```

```python
car = Car(BongoEngine())
car.drive()
```

```python
# 강결합을 의도적으로 구현

# 람보르기니니까 람보르기니 엔진만 들어가야하는게 맞잖ㄴ아요..
class Lambo(Car):
  def __init__(self):
    super(Lambo, self).__init__(LamboEngine())
```

```python
lambo = Lambo()
lambo.drive()
```

```python
# 봉고니까 봉고 엔지만 들어가야 하는게 맞잖아요.
class Bongo(Car):
  def __init__(self):
    super(Bongo, self).__init__(BongoEngine())

bongo = Bongo()
bongo.drive()
```

## use a 관계
- A use a B : A가 B를 사용한다.
- 부품은 아니지만 A 클래스의 특정 ***메소드***를 수행할 때 B객체를 넣어야 하는 관계

```python
class Car:
  def drive_through(self, food):
    print("드라이브 쓰루하면서")
    food.eat()
```

```python
class Hamburger:
  def eat(self):
    print("참깨빵 위에~~~~특별한 소스 양상추까지~ 빨빠빠밤")
```

```python
h = Hamburger()
c = Car()

# 자동차로 드라이브 쓰루를 하면서 햄버거를 먹는것
c.drive_through(h)
```
