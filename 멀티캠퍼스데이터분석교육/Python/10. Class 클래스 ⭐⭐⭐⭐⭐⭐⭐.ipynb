```python
class User4:

  def __init__(self, first_name, age):

    # setterë¥¼ ì´ìš©í•´ì„œ ì´ˆê¸°í™”
    self.set_first_name(first_name)  #self.first_name = first_nameì´ë ‡ê²Œ í•´ë„ë¨
    self.age = age   #self.set_age(age) í•´ë„ë¨

  # first_nameì— ëŒ€í•œ setter
  def set_first_name(self, first_name):

    if len(first_name) >= 3:
      self.__first_name = first_name
    else:
      self.__first_name = None

  # ageì— ëŒ€í•œ setter
  def set_age(self, age):

    if age > 0 :
      self.__age = age
    else:
      self.__age = None

  # first_nameì— ëŒ€í•œ getter
  def get_first_name(self):
    # ì´ë¦„ ì¤‘ê°„ì˜ ëª¨ë“  ê¸€ìê°€ *ë¡œ í‘œì‹œ ë  ìˆ˜ ìˆë„ë¡
    # minho -> m***o
    first_char = self.__first_name[0]
    last_char  = self.__first_name[-1]
    star_len   = len(self.__first_name)-2
    stars      = "*" * star_len

    return f"{first_char}{stars}{last_char}"

  # ageì— ëŒ€í•œ getter
  def get_age(self):
    return self.__age

  # ì‚¬ëŒì˜ ì •ë³´ë¥¼ í‘œê¸°í•˜ê¸° ìœ„í•œ ë¬¸ìì—´ì„ ë§Œë“œëŠ” ë©”ì†Œë“œë¥¼ êµ¬í˜„
  def __make_user_info(self):
    return f"{self.get_first_name()}ì˜ ë‚˜ì´ëŠ” {self.get_age()}"

  def print_user_info(self):
    user_info_text = self.__make_user_info()
    print("â­", user_info_text, "ğŸ˜Š")

  # getter & setter ë“±ë¡
  first_name = property(get_first_name, set_first_name)
  age = property(get_age, set_age)
```
