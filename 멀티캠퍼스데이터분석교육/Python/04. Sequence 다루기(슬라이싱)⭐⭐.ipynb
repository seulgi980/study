# 슬라이싱(Slicing) ⭐⭐👍⭐✌❤❤🤣

내가 원하는 만큼 잘라 가져오기

잘라서 새로운 시퀀스 만들어 낸다.

[시작 : 끝 : step]
***끝 -1*** 까지 자름!!⭐

```python
text = "안녕하세요!반갑습니다"
print(text)
```

```python
text[0 : 3]
```

```python
text[5: 8]
```

```python
text[5: 500]
# slice는 엔드 끝제한이 없음 걍 자동으로 마지막까지 추출함
```

```python
text[3 : -2]
# 음수 양수 인덱스 혼용가능
```

```python
text[-11 : 10]
```

```python
text[7 : -10]
# 역주행 불가!!
```

start end step생략가능

start 안쓰면 처음부터 0 으로세고

end 안쓰면 시퀀스 길이 len가들어가고

step 안쓰면 1이 들어감

```python
text[ : -5]
```

```python
file_name = "dfsfdsfdfsfsdfsfd.jpg"
file_name[ :-4]
```

# step에 대한 이야기
- 지정한 숫자만큼 ***건너 뛰면서*** 슬라이싱
- 생략하면 기본적으로 1이 주어짐
- 부호를 이용해서 방향 설정 가능
스텝이 음수면 역방향으로 뽑는다.

```python
text = "안녕하세요!반갑습니다"
```

```python
text [ : : 2]
```

```python
text[ : :-2]
```

```python
text [ : :3]
```

```python
text[3 : 8 :2]
```

```python
text[ : : -1]
# 역방향 출력
```

```python
text[8 : 3 :-2]
# 역방향일땐 end +1이됨! 내가 맞췄더ㅏㅇ 히힣
```

```python
text[-3 : -8: -2]
```

```python
text[-2 : -6 :-2]
# end 하나 앞까지 세어야하는데 실수해따
```
