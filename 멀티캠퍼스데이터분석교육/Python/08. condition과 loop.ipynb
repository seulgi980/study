# 반복문 loop
- 'while' : 조건제 반복(~동안에)
  - ~라면 동안 코드컨텍스트를 반복적으로 실행해라
- ⭐'for ~ in' : 컬렉션 반복
  - 순회 가능한 컬렉션을 더 이상 순회 할 수 없을 때 까지 반복
  
  순회 = iterable : 컬렉션 내의 item을 하나씩 접근할 수 있다.

  컬렉션 = 리스트, 튜플, 셋, 딕셔너리, str문자열

- iterator : 순회를 가능하게 하는 개념

```python
data = 3

while data != 0:   #if랑 문법 비슷 : 빼지말기!!!   데이터가 0 이 아닌 동안 반복해라
  print("Hello", data)
  data -= 1

print("done")
```

```python
point_lst = [80, 90, 100]
# index쓰는방법

총점 = 0
data_len = len(point_lst)
i = 0

while i < data_len:
  총점 += point_lst[i]
  i += 1

print(f"합계 = {총점}")
print(f"평균 = {총점 / len(point_lst)}")

```

```python
# pop을 활용한 방식
총점 = 0

while len(point_lst) != 0:
  point = point_lst.pop()
  총점 += point

print(f"합계 : {총점}")
print(f"평균 : {총점 / 3}")
```

```python
#내가 복습
point_lst = [80, 90, 100]

총점 = 0

while len(point_lst) != 0:
  총점 += point_lst.pop() #⭐⭐.pop()

print(f"합계 : {총점}")
print(f"평균 : {총점 / 3}")
```

# zip 함수 사용하기 ⭐⭐⭐⭐
여러개의 컬렉션을 하나로 모아서 병렬로 순회가 가능하게 함

갯수제한x


```python
student_name = ["A", "B", "C"]
score_list = [90, 87, 93]

for name, score in zip(student_name, score_list):
  print(f"{name}의 점수는 {score}점 입니다.")   # zip이뭔가 패킹같으니 언패킹도 가능!
```

```python
# 그럼 컬렉션의 크기가 다르면 우짜농?
#⭐⭐⭐⭐크기가 작은 것 만큼만 반복!!

lst1 = [1,2,3,4]
lst2 = ["a","b","c"]

for 숫자, 영어 in zip(lst1, lst2):
  print(숫자, 영어)
  print(f"{숫자}고요 {영어}예요")
```

# enumerate

for  인덱스, 데이터 in enumerator

변수 2개!

```python
lst = ["A", "B", "C"]

for idx, data in enumerate(lst):
  print(idx, data)
```

```python
lst1 = ["A", "B", "C"]
lst2 = [10,20,30]

for idx, (item1, item2) in enumerate(zip(lst1, lst2)):
  print(idx, item1, item2)
```

```python
lst1 = ["A", "B", "C"]
lst2 = [10,20,30]
lst3 = [-1,-2,-3]

for idx, (item1, item2, item3) in enumerate(zip(lst1, lst2, lst3)):
  print(idx, item1, item2, item3)
```

# 컴프리헨션 😃😃😃⭐⭐⭐
- list set dict를 간단하게 만들 수 있는 문법
- 튜플은 컴프리헨션으로 만들 수 없음!

반복문을 이용해서 컬렉션에 데이터를 추가하는 경우가 많다.

```python
# 1~5 까지 각각의 정수에 제곱한 결과를 리스트에 담아보기!
result = []

for data in range(1,6):
  result.append(data ** 2)  #**2는 제곱 ㅎ ㅎ **3세제곱!

print(result)
```

```python
2 **5
```

list 컴프리헨션

'[<**expr**> for <**value**> in <**collection**> if <**condition**>]'


if 문은 있어도  되고 없어도 된다.
<추가할 값에 대한 처리> data ** 2 같은 부분!

조건이 True일 때 콜렉션에 아이템들을 리스트에 추가해준다!

```python
result = [data**2 for data in range(1,6)]  #무조건 for data in range(1,6)부터 쓰기 무조건!!⭐⭐
print(result)
```

```python
# 1~100 까지 홀수인 정수에 대해서 제곱한 후 추가!!

result = [data**2 for data in range(1,101) if data % 2 == 1]
print(result)
```

컴프리헨션이 for문보다 빠르기 때문에 배우는겨!!

```python
# 1~30 까지의 숫자에서 홀수는 제곱을하고 짝수는 -1을 곱해서 추가하는데, 3의 배수는 제외하고 추가!
result = [data**2 if data % 2 == 1 else data * -1 for data in range(1, 31) if data % 3 != 0]
print(result)
#위에서 배운 삼항연산 이용!
```

```python
# 30에서 60까지의 숫자를 넣는데 짝수는 /2를 하고 홀수는 정수로 넣는다 하지만 10의 배수는 제외 한다
result = [data / 2 if data % 2 ==0 else data for data in range(30,61) if data % 10 != 0]
print(result)
```
